```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import np_utils
from keras.datasets import mnist
%matplotlib inline
```

```python
from sklearn.model_selection import train_test_split
frame = pd.read_csv('10x10, 10^5 images')
X = frame.drop('100', axis = 1).values
y = frame['100'].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
np.random.seed(42)
```

## Создаём последовательную модель №1

```python
model = Sequential()
#Добавляем уровни сети
model.add(Dense(1000, input_dim=100, kernel_initializer='normal', activation='relu'))
model.add(Dense(50, kernel_initializer='normal', activation='sigmoid'))
model.add(Dense(1, kernel_initializer='normal', activation='linear'))
```

```python
#Компилируем модель
model.compile(loss='mean_squared_error', optimizer='SGD', metrics=['mean_squared_error'])
print(model.summary())
```

```python
%%time
#Обучаем сеть
model.fit(X_train, y_train, batch_size=100, epochs=20, verbose=True)
```

```python
from sklearn.metrics import mean_squared_error
print(mean_squared_error(y_train, model.predict(X_train)))
print(mean_squared_error(y_test, model.predict(X_test)))
```
