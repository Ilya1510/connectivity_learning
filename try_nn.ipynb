```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import np_utils
from keras.datasets import mnist
%matplotlib inline
```

```python
from sklearn.model_selection import train_test_split
frame = pd.read_csv('10x10, 10^5 images')
X = frame.drop('100', axis = 1).values
y = frame['100'].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
from sklearn.metrics import mean_squared_error
np.random.seed(42)
```

# Создаём последовательную модель №0

```python
model = Sequential()
#Добавляем уровни сети
model.add(Dense(400, input_dim=100, kernel_initializer='normal', activation='relu'))
model.add(Dropout(0.25))
model.add(Dense(20, kernel_initializer='normal', activation='sigmoid'))
model.add(Dense(1, kernel_initializer='normal', activation='linear'))
```

```python
#Компилируем модель
model.compile(loss='mean_squared_error', optimizer='SGD', metrics=['mean_squared_error'])
print(model.summary())
```

# Convolutional network

```python
import numpy as np
from keras.datasets import cifar10
from keras.models import Sequential
from keras.layers import Dense, Flatten, Activation, Dropout
from keras.layers.convolutional import Conv2D, MaxPooling2D
from keras.utils import np_utils
from sklearn.metrics import mean_squared_error
```

```python
model = Sequential()
#Первый свёрточный слой
model.add(Conv2D(20, (2, 2), padding='same', input_shape=(10, 10, 1), activation='relu'))
#Второй свёрточный слой
model.add(Conv2D(20, (2, 2), activation='relu'))
#Слой подвыборки
model.add(MaxPooling2D(pool_size=(2, 2)))
#Слой регуляризации
model.add(Dropout(0.25))
#Третий свёрточный слой
model.add(Conv2D(20, (2, 2), padding='same', activation='relu'))
#Четвёртый свёрточный слой
model.add(Conv2D(20, (2, 2), activation='relu'))
#Второй слой подвыборки
model.add(MaxPooling2D(pool_size=(2, 2)))
#Слой регуляризации
model.add(Dropout(0.25))
#Преобразование из двумерного вида в плоский
model.add(Flatten())
#Полносвязный слой
model.add(Dense(256, activation='relu'))
#Слой регуляризации
model.add(Dense(32, activation='relu'))
#Выходной слой
model.add(Dense(1, activation='linear'))
```

```python
#Компилируем сеть
model.compile(loss='mean_squared_error', optimizer="SGD", metrics=['mean_squared_error'])
```

```python
model.summary()
```

```python
%%time
model.fit(X_train.reshape((80000, 10, 10, 1)), y_train, batch_size=100, epochs=5, validation_split=0.1, shuffle=True, verbose=True)
print('mse train', mean_squared_error(y_train, model.predict(X_train.reshape((80000, 10, 10, 1)))))
print('mse test', mean_squared_error(y_test, model.predict(X_test.reshape((20000, 10, 10, 1)))))
```
